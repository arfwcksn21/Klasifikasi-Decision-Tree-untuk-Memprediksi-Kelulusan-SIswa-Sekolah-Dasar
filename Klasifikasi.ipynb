```python
import pandas as pd
```

```python
df = pd.read_excel('SD_kec_juai.xlsx')
df
```

```python
df.isna().sum()
```

```python
df.Jumlah.fillna(0, inplace=True)
df.Rerata_Keseluruhan.fillna(0, inplace=True)
```

```python
df.head(50)
```

```python
df.isna().sum()
```

```python
d_train = df[['NISN','mean_PA','mean_PKn','mean_Matematika','mean_IPA','mean_IPS','mean_SBK','mean_Penjaskes','mean_Bataqu','mean_BSB','Jumlah','Rerata_Keseluruhan']].reset_index(drop=True)
d_train.fillna(0, inplace=True)
d_train
```

```python
#import library yang akan digunakan
from IPython.display import Image
from sklearn.impute import SimpleImputer
from sklearn import tree
from subprocess import check_call
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import numpy as np
```

```python
for index, row in d_train.iterrows():
    if row['Rerata_Keseluruhan'] >=70:
        d_train.loc[index, 'Status'] = 'Lulus'
    else:
        d_train.loc[index, 'Status'] = 'Tidak Lulus'
d_train
```

```python
y = d_train['Status'].values
y
```

```python
columns = ['mean_PA','mean_PKn','mean_Matematika','mean_IPA','mean_IPS','mean_SBK','mean_Penjaskes','mean_BSB']
features = d_train[list(columns)].values
features
```

```python
imputer = SimpleImputer(missing_values= np.NAN, strategy= 'mean', fill_value=None, verbose=0, copy=True)
x = imputer.fit_transform(features)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(x_train, y_train)
predict = clf.predict(x_test)
```

```python
with open("Graph_tree.dot", 'w') as f:
    f = tree.export_graphviz(clf, out_file=f, feature_names = columns, filled=True, rounded=True, special_characters = True)
```

```python
check_call(['dot', '-Tpng', 'Graph_tree.dot', '-o', 'Graph_tree.png'])
```

```python
Image(filename='Graph_tree.png')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(5,3))
sns.heatmap(confusion_matrix(y_test, predict), annot=True, fmt='.0f', ax=ax)
plt.xlabel("predict")
plt.ylabel("actual")

plt.show
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predict))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predict)
```
